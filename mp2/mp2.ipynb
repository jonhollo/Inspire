```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt
```

```python
img = cv.imread('mcgtx.jpg') #takes in the image
cv.imshow('MCGTX', img)

cv.waitKey(0)
cv.destroyWindow('MCGTX')
```

```python
imgGS = cv.cvtColor(img, cv.COLOR_BGR2GRAY) #applies grayscale filter to 

cv.imshow('MCGTX', imgGS)

gsT = cv.transpose(imgGS) #transposes grayscaled image

cv.waitKey(0)
cv.destroyWindow('MCGTX')

cv.imshow('MCGTX', gsT)

cv.waitKey(0)
cv.destroyWindow('MCGTX')

imgGS = imgGS/255
```
